```python
# Import required libraries 
import sympy
import time
```

```python
# Generate prime numbers 
#primes=[]
#start = time.time()
primes=list(sympy.primerange(100000, 1000000))
#end = time. time()
#print(end - start)
```

```python
len(primes)
```

```python
checked=[]
for num in primes:
    num=str(num)
    # using set() + count() to get count of each element in string 
    res = {i : num.count(i) for i in set(num)}
    if max(res.values()) ==3:
        checked.append(num)
```

```python
len(checked)
```

```python
checked=list(map(int,checked))
```

```python
checked[888]
```

```python
# find smallest eight prime family member 

et_prime=[]
for num in checked:
    #c+=1
    num=str(num)
    #print(num)
    # using set() + count() to get count of each element in string 
    res = {i : num.count(i) for i in set(num)}
    
    #find digit to be replaced
    rep=list(res.keys())[list(res.values()).index(max(res.values()))]
    #print(rep)
    # generate new lists by replacing digits
    ep=[]
    n=num
    for i in range(10):  
        n=n.replace(rep,str(i))
        ep.append(sympy.isprime(int(n)))
        n=num
    if sum(ep)==8:
        et_prime.append(num)
        
    
    #if max(res.values()) >=3:
    #    checked.append(num)
```
